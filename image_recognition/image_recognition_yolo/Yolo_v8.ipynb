# Train the model

```python
import torch
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8n.yaml').load('yolov8n.pt')  # build from YAML and transfer weights

# Use the model
model.train(data="/content/data.yaml", epochs=50, imgsz=640)  # train the model
metrics = model.val()  # evaluate model performance on the validation set


```

# Evaluate training

```python
from PIL import Image


results = model("left0002.jpg")  # predict on an image
# Show the results
for r in results:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.show()  # show image
    im.save('results_cutlery.jpg')  # save image
```
