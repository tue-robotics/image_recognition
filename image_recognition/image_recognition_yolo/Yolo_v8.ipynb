# Train the model

```python
from ultralytics import YOLO
# Load a model
model = YOLO('yolov8n.pt')  # build from YAML and transfer weights

```

```python
#model = YOLO('/content/runs/detect/train8/weights/last.pt') #Resume training bc validation loss still decreasing --> resume=True
# Use the model
model.train(data="/content/cutlery_31okt2023/data.yaml", epochs=100, patience=20, freeze=22, batch=128, imgsz=(480, 640))  # train the model

```

```python
metrics = model.val()  # evaluate model performance on the validation set
```

# Evaluate training

```python
import os
from PIL import Image

test_directory = "/content/cutlery_31okt2023/test/images/"
results_directory = "/content/results/"

if not os.path.exists(results_directory):
    os.makedirs(results_directory)

test_images = os.listdir(test_directory)


for image_filename in test_images:
    image_path = os.path.join(test_directory, image_filename)

    results = model(image_path)

    # Show and save the results
    for idx, r in enumerate(results):
        print(r.boxes.conf)
        print(r.boxes.cls)
        im_array = r.plot()  # plot a BGR numpy array of predictions
        im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
        result_filename = f"{os.path.splitext(image_filename)[0]}_result_{idx}.jpg"
        result_path = os.path.join(results_directory, result_filename)
        im.save(result_path)  # save image
```
